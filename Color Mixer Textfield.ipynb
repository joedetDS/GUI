```python
import tkinter as tk

def mix_color():
	
	red_density=float(redEntry.get())
	blue_density=float(blueEntry.get())
	green_density=float(greenEntry.get())
	
	red=int(red_density)
	green=int(green_density)
	blue=int(blue_density)
    
	color=f"#{red:02x}{green:02x}{blue:02x}"
    
	root.config(bg=color)
	mix.config(bg=color)
	mix1.config(bg=color)
	mix2.config(bg=color)
	mix3.config(bg=color)
	mix0.config(bg=color)
    
    

root=tk.Tk()
root.title("Color Mixer")

mix=tk.Label(root,height=15)
mix.pack()


titleLabel=tk.Label(root,bg="white",width=25,text="COLOR MIXER",height=2,font='Arial 20 bold')
titleLabel.pack()
mix0=tk.Label(root,height=1)
mix0.pack()	

redLabel=tk.Label(root,text="RED: ",bg='red',fg='white')
redLabel.pack()	
redEntry=tk.Entry(root)
redEntry.pack()
mix1=tk.Label(root,height=1)
mix1.pack()

blueLabel=tk.Label(root,text="BLUE: ",bg='blue',fg='white')
blueLabel.pack()	
blueEntry=tk.Entry(root)
blueEntry.pack()
mix2=tk.Label(root,height=1)
mix2.pack()

greenLabel=tk.Label(root,text="GREEN: ",bg='green',fg='white')
greenLabel.pack()	
greenEntry=tk.Entry(root)
greenEntry.pack()
mix3=tk.Label(root,height=1)
mix3.pack()

mixbutton=tk.Button(root,text="Mix",command=mix_color,width=18)
mixbutton.pack()

root.mainloop()
```
